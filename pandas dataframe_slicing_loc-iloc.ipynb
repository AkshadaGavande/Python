# Slicing the dataframe
 We can select specific ranges of our data in both the row and column directions using either label or integer-based indexing.

- loc: indexing via labels or integers
- iloc: indexing via integers

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/i life/Desktop/python basics/mtns.csv",names=['name','height','mountain range','summited'])
df
```

```python
# select the first, second and third rows
df[:3]
```

```python
# select the last row in the dataframe
df[-1:]
```

```python
# select columns
df[['name','height']]
```

```python
#select colum by name
df.loc[:,'summited']
```

```python
#get numpy array 
df.loc[:,'summited'].values
```

```python
#select several consecutive columns

df.loc[:, 'height': 'summited']
```

```python
#
df.loc[:, ['height', 'summited']]
```

```python

print([df.loc[:, 'summited']])

#select rows based on column values

df.loc[df.loc[:, 'summited'] > 1954, :]
```

- To select data by its position, we use the .iloc[] function. 

- The first argument is for the rows, and the second argument is for the columns. 

```python
# select column by position
df.iloc[:,2]
```

```python
#select row and column values by
df.iloc[0,2]
```

```python
#select several rows by position
df.iloc[[2,3],:]
```

```python
# select several adjacent columns by position
df.iloc[:,0:2]
```

```python
# get numpy array
df.iloc[:,[1,3]].values
```
